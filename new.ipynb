```python
def getSoup(link):
    sleep(0.1) #to be kind to the website
    r = requests.get(link)
    r.encoding = 'UTF-8'
    return BeautifulSoup(r.text,'lxml')
```

```python
soup = getSoup('https://carvago.com/cs/auta/skoda')
```

```python
tds = soup.findAll('div', {'class':'CarCard__CarCardItem-sc-tt14yf-3 bSEhwY'})
#tds
```

```python
def getAllLinks(link):
    soup = getSoup(link)
    tds = soup.findAll('div', {'class':'CarCard__CarCardItem-sc-tt14yf-3 bSEhwY'})
    return ['https://carvago.com' + td.find('a')['href'] for td in tds]

links = getAllLinks('https://carvago.com/cs/auta/skoda')
links
```

```python
def getName(link):
    soup = getSoup(link)
    return soup.find('h1',{'class':'Text-sc-1wbtj74-0 kOImJd DetailTitleInfo__StyledH1-sc-wvwau1-5 cVndGo'}).text.strip()

name = [getName(link) for link in links]
name
```

```python
def getPrice(link):
    soup = getSoup(link)
    return soup.find('div',{'class':'HeaderPrice__PrimaryPrice-sc-jl5zco-1 jpZNng'}).text.strip()

price = [getPrice(link) for link in links]
price
```

```python
def getSeller(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'BaseInfo__BaseInfoValue-sc-12khx42-3 jqQUus'})[3].text.strip()

seller = [getSeller(link) for link in links]
seller
```

```python
def getCountry(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'BaseInfo__BaseInfoValue-sc-12khx42-3 jqQUus'})[4].text.strip()

country = [getCountry(link) for link in links]
country
```
