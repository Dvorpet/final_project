```python
def getSoup(link):
    sleep(0.1) #to be kind to the website
    r = requests.get(link)
    r.encoding = 'UTF-8'
    return BeautifulSoup(r.text,'lxml')
```

```python
soup = getSoup('https://carvago.com/cs/auta/skoda')
```

```python
divs = soup.findAll('div', {'class':'css-1dapwlk e1oahio83'})
divs
```

```python
def getAllLinks(link):
    soup = getSoup(link)
    tds = soup.findAll('div', {'class':'css-1dapwlk e1oahio83'})
    return ['https://carvago.com' + div.find('a')['href'] for div in divs]

links = getAllLinks('https://carvago.com/cs/auta/skoda')
links
```

```python
def getName(link):
    soup = getSoup(link)
    return soup.find('h1',{'class':'ea6i44d0 css-13i2d41 e5xkr5t0'}).text.strip()

name = [getName(link) for link in links]
name
```

```python
def getPrice(link):
    soup = getSoup(link)
    return soup.find('div',{'class':'css-1mbrvie e1hgzarh2'}).text.strip()

price = [getPrice(link) for link in links]
price
```

```python
def getSeller(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[3].text.strip()

seller = [getSeller(link) for link in links]
seller
```

```python
def getCountry(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[4].text.strip()

country = [getCountry(link) for link in links]
country
```

```python
def getMileage(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[5].text.strip()

mileage = [getMileage(link) for link in links]
mileage
```

```python
def getRegistration(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[6].text.strip()

registration = [getRegistration(link) for link in links]
registration
```

```python
def getTransmission(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[7].text.strip()

transmission = [getTransmission(link) for link in links]
transmission
```

```python
def getPower(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[8].text.strip()

power = [getPower(link) for link in links]
power
```

```python
def getFueltype(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[9].text.strip()

fuel_type = [getFueltype(link) for link in links]
fuel_type
```

```python
def getConsumption(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[10].text.strip()

consumption = [getConsumption(link) for link in links]
consumption
```

```python
def getDrivetrain(link):
    soup = getSoup(link)
    return soup.find_all('div',{'class':'css-o5n6vn e51cp1p1'})[11].text.strip()

drivetrain = [getDrivetrain(link) for link in links]
drivetrain
```

```python
df = pd.DataFrame(list(zip(links,name,price,seller,country,mileage,registration,transmission,power,fuel_type,consumption,drivetrain)))
```

```python
df.columns = ['links','name','price','seller','country','mileage','registration','transmission','power','fuel_type','consumption','drivetrain']
```
